```python
#Function Definition
def header(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    header1=[]
    header2=[]
   
    for i in soup.find_all('span',class_="central-textlogo__image sprite svg-Wikipedia_wordmark"):
        header1.append(i.text.split('\n')[1])
    print('\n Header1:',header1)
    
    for i in soup.find_all('strong',class_="jsl10n localized-slogan"):
        header2.append(i.text)
    print('\n Header2:',header2)
    
#Calling Function
header('https://www.wikipedia.org/')
```

# 3) Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame.

```python
#Function Definition
def top_100_hindi_movies(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    name=[]
    rating=[]
    year_of_release=[]
  
    for i in soup.find_all('td',class_="titleColumn"):
        name.append(i.text.split('\n')[2])
    
  
    for i in soup.find_all('td',class_="ratingColumn imdbRating"):
        rating.append(i.text.split('\n')[1])
    
    
    for i in soup.find_all('span',class_="secondaryInfo"):
        year_of_release.append(i.text)
   
    
    df=pd.DataFrame({
        'Names':name,
        'Ratings':rating,
        'year_of_release':year_of_release
    })
    return(df[0:101])
    
#Calling Function   
top_100_hindi_movies('https://www.imdb.com/india/top-rated-indian-movies/?sort=rk,asc&mode=simple&page=1')
```

# 4) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) 
from https://presidentofindia.nic.in/former-presidents.htm

```python
#Function Definition
def presidents_of_india(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    name=[]
    term_of_office=[]
  
    for i in soup.find_all('div',class_="presidentListing"):
        name.append(i.text.split('\n')[1])
    
  
    for i in soup.find_all('div',class_="presidentListing"):
        term_of_office.append(i.text.split('\n')[2])
    
    df=pd.DataFrame({
        'Names':name,
        'Term_of_office':term_of_office
       })
    return(df)
    
#Calling Function   
presidents_of_india('https://presidentofindia.nic.in/former-presidents.htm')
```

# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
#Function Definition
def Top_10_Men_ODI_team(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    teams=[]
    matches=[]
    points=[]
    ratings=[]
    for i in soup.find_all('span',class_="u-hide-phablet"):
        teams.append(i.text)
    matches.append(soup.find('td',class_="rankings-block__banner--matches").text)
    points.append(soup.find('td',class_="rankings-block__banner--points").text)
    ratings.append(soup.find('td',class_="rankings-block__banner--rating u-text-right").text.split('\n')[1])
    count=0
    for i in list(soup.find_all('td',class_="table-body__cell u-center-text"))[:18]:
        if count%2==0:
            matches.append(i.text)
        else:
            points.append(i.text)
        count+=1
    for i in list(soup.find_all('td',class_="table-body__cell u-text-right rating"))[:9]:
        ratings.append(i.text)
    df=pd.DataFrame({
        'Teams':teams[:10],
    'Matches':matches,
    'Points':points,
    'Ratings':ratings})
    return(df)
    
#calling Function
Top_10_Men_ODI_team('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

# 7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world
i) Headline
ii) Time
iii) News Link

```python
#Function Definition
def news(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    headline=[]
    time=[]
    news_link=[]
    
    
    for i in list(soup.find_all('div',class_="RiverHeadline-headline RiverHeadline-hasThumbnail"))[:4]:
        headline.append(i.text)
    
 
    for i in list(soup.find_all('span',class_="RiverByline-datePublished"))[:4]:
        time.append(i.text)
        
    for div in soup.findAll('div',{'class':'RiverHeadline-headline RiverHeadline-hasThumbnail'})[:3]:
        a = div.findAll('a')
        for link in a:
            news_link.append(link.get('href'))
          
        
   
        
    print(len(headline))
    print(len(time))
    print(len(news_link))
        
   
        
    df=pd.DataFrame({
        'Headline':headline,
        'time':time,
        'news link':news_link
       
        
    })
    print(df)
    
#Calling Function   
news('https://www.cnbc.com/world/?region=world')
```

# 8) Write a python program to scrape the details of most downloaded articles from AI in last 90 days. 
https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details :
i) Paper Title 
ii) Authors
iii) Published Date 
iv) Paper URL

```python
#Function Definition
def article(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    paper_title=[]
    authors=[]
    published_date=[]
    paper_url=[]
    
    for i in soup.find_all('h2',class_='sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR'):
        paper_title.append(i.text)
    
  
    for i in soup.find_all('span',class_="sc-1w3fpd7-0 pgLAT"):
        authors.append(i.text)
    
    
    for i in soup.find_all('span',class_="sc-1thf9ly-2 bKddwo"):
        published_date.append(i.text)
   
    
    for i in soup.find_all('a',class_="sc-5smygv-0 nrDZj"):
        paper_url.append(i.get('href'))
        
    df=pd.DataFrame({
        'Paper Title':paper_title,
        'Authors':authors,
        'Published Date':published_date,
        'paper_url':paper_url
    })
    return(df)
   

 #Calling Function 
article('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location 
iv) Ratings
v) Image URL

```python
#Function Definition
def dineout(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    restaurant_name=[]
    cuisine=[]
    location=[]
    ratings=[]
    image_url=[]
    
    for i in soup.find_all('a',class_="restnt-name ellipsis"):
        restaurant_name.append(i.text)
        
    for i in soup.find_all('span',class_="double-line-ellipsis"):
        cuisine.append(i.text.split('|')[1])
        
    for i in soup.find_all('div',class_="restnt-loc ellipsis"):
        location.append(i.text)
        
    for i in soup.find_all('div',class_="restnt-rating rating-4"):
        ratings.append(i.text)
        
    for i in soup.find_all('img',class_="no-img"):
        image_url.append(i.get('data-src'))
        
    df=pd.DataFrame({
        'restaurant names': restaurant_name,
        'cuisine':cuisine,
        'location':location,
        'ratings':ratings,
        'image_url':image_url
    })
    return(df)
    
#calling Function
dineout('https://www.dineout.co.in/delhi-restaurants/welcome-back')
```

# 10) Write a python program to scrape the details of top publications from Google Scholar from 
https://scholar.google.com/citations?view_op=top_venues&hl=en
i) Rank 
ii) Publication
iii) h5-index
 iv) h5-median

```python
#Function Definition
def google_scholar(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    rank=[]
    publication=[]
    h5_index=[]
    h5_median=[]
    
    for i in soup.find_all('td',class_="gsc_mvt_p"):
        rank.append(i.text)
    
  
    for i in soup.find_all('td',class_="gsc_mvt_t"):
        publication.append(i.text)
    
    count=0
    for i in soup.find_all('td',class_="gsc_mvt_n")[:202]:
        if count%2==0:
            h5_index.append(i.text)
        else:
            h5_median.append(i.text)
        count+=1
   
    df=pd.DataFrame({
        'rank':rank,
        'publication':publication,
        'h5 index':h5_index,
        'h5 median':h5_median
    })
    return(df)
    
#calling Function
google_scholar('https://scholar.google.com/citations?view_op=top_venues&hl=en')
```
