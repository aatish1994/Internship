```python
#scrapping airline name
airline_name=[]
try:
    airline_name_tag=driver.find_elements(By.XPATH,'//span[@class="codeshares-airline-names"]')
    for j in airline_name_tag:
        airline_name.append(j.text)
except NoSuchElementException : #handling no such element exception
    airline_name.append('-')  
```
